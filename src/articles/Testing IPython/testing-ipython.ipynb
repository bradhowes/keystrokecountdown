```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps)) * 2 - 1
draws
```

```python
walk = np.cumsum(draws, 1)
walk
```

```python
x = np.arange(0, nsteps)
for i in range(0,10):
    plt.plot(x, walk[i])
```
